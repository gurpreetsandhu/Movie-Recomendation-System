```python
path = 'the-movies-dataset'
filename = 'keywords.csv'
file = []
with open(path + '/' + filename, 'r', errors = 'ignore') as csvfile:
    spamreader = csv.DictReader(csvfile)
    for row in spamreader:
        file.append(row)
```

```python
keywords_file = []
for i in range(len(file)):
    keywords_file.append([int(file[i]['id']), file[i]['keywords']])
```

```python
import pickle

with open('Keyword', 'wb') as fp:
    pickle.dump(keywords_file, fp)
```

```python
def get_id_and_key(text):
    ret_data = []
    for i in range(len(text)):
        key_word_list = ast.literal_eval(text[i][1])
        id_n = int(text[i][0])
        ret_data.append([id_n, key_word_list])
    return ret_data
```

```python
keywords = get_id_and_key(keywords_file)
```

```python
def get_only_keywords_for_w2v(kw):
    docs = []
    stop_words = list(stopwords.words('english'))
    tokenizer = RegexpTokenizer('[a-zA-Z0-9]*[*.*]?\w+')
    for i in range(len(kw)):
        if len(kw[i][1]) > 0:
            temp = []
            for j in range(len(kw[i][1])):
                tokens = tokenizer.tokenize(kw[i][1][j]['name'])
                for k in tokens:
                    if k not in stop_words:
                        temp.append(k)
            if len(temp) > 0:
                docs.append(temp)
    return docs
```

```python
only_keywords_doc = get_only_keywords_for_w2v(keywords)
```

```python
only_keywords_doc[0]
```

```python
import gensim
```

```python
model = gensim.models.Word2Vec(only_keywords_doc, size=150, window=10, min_count=2, workers=10)
        
model.train(only_keywords_doc, total_examples=len(only_keywords_doc), epochs=10)
```

```python
ex = "toy"
model.wv.most_similar(positive = ex)
```

```python
my_model = model.wv.vocab
```

```python
def get_only_overview_for_w2v(text):
    docs = []
    stop_words = list(stopwords.words('english'))
    tokenizer = RegexpTokenizer('[a-zA-Z0-9]*[*.*]?\w+')
    for i in range(len(text)):
        tokens = tokenizer.tokenize(text[i][1])
        temp_tk = []
        for j in tokens:
            if j not in stop_words:
                temp_tk.append(j)
        docs.append(temp_tk)
    return docs
```

```python
only_overview = get_only_overview_for_w2v(movies_overview)
```

```python
model_2 = gensim.models.Word2Vec(only_overview, size=150, window=10, min_count=2, workers=10)
        
model_2.train(only_overview, total_examples=len(only_overview), epochs=10)
```

```python
ex = "toy"
model_2.wv.most_similar(positive = ex)
```
